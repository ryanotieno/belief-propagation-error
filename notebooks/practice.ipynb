```python
from jax import grad

def sum_logistic(x):
  #jnp.exp will go ahead and produce an array of e^x[n] of size x.len(), and 1 + just adds to all the elements of the array
  #jnp.sum collapses them into a single value
  return jnp.sum(1.0 / (1.0 + jnp.exp(-x)))

x_small = jnp.arange(3.)
#getting called in a smilar way to jit
derivative_fn = grad(sum_logistic)
print(derivative_fn(x_small))
```

```python
from jax import random

key = random.key(1701)
print(key)
#Defaults to 2 as shape/keys generated from this operation
key1, key2 = random.split(key)
print(key1, key2)
#as opposed to random.rand, this will give me values within a certain bound. like -4 < x < 4
#Returns A normal random array with the specified shape and dtype. 150 rows, 100 cols
mat = random.normal(key1, (150, 100))
batched_x = random.normal(key2, (10,100))
def apply_matrix(x):
  return jnp.dot(mat, x)
#can only do jnp.dot when I give it one row of my batched_x matrix.
print(apply_matrix(batched_x[0]))
```

```python
def naively_batched_apply_matrix(v_batched):
  #stack allows the jnp.dot to receive one row of v_batched at a time, and then add it back to a matrix once the dot product is computed.
  return jnp.stack([apply_matrix(v) for v in v_batched])

print('Naively batched')
%timeit naively_batched_apply_matrix(batched_x).block_until_ready()
```

```python
import numpy as np

@jit
def batched_apply_matrix(batched_x):
  return jnp.dot(batched_x, mat.T)
#why doesnt this need stack? is it cause we take the 100 cols of matrix x and the 100 cols of mat, and because mat is transposed, its now 100 rows?
np.testing.assert_allclose(naively_batched_apply_matrix(batched_x),
                           batched_apply_matrix(batched_x), atol=1E-4, rtol=1E-4)
print('Manually batched')
%timeit batched_apply_matrix(batched_x).block_until_ready()
```

```python
from jax import vmap

@jit
def vmap_batched_apply_matrix(batched_x):
  return vmap(apply_matrix)(batched_x)

np.testing.assert_allclose(naively_batched_apply_matrix(batched_x),
                           vmap_batched_apply_matrix(batched_x), atol=1E-4, rtol=1E-4)
print('Auto-vectorized with vmap')
%timeit vmap_batched_apply_matrix(batched_x).block_until_ready()
#additional run time is probably because it is doing some thinking as well of whether a stack is needed or not.
#but overall it is better because I dont need to know what the shapes are when trying to compute dot product
```
