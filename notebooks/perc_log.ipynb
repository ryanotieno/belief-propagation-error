```python
N = 10 #no. of nodes
p = 0.4 #ocu
G = nx.erdos_renyi_graph(N, p)
A = nx.to_numpy_array(G)
pos = nx.spring_layout(G)
labels = {node: str(node) for node in G.nodes()}
```

```python
for i in range(N):
    for j in range(N):
        temp_list = []
        for k in range(N):
            if A[j][k] == 1 and k != i:
                temp_list.append(1)
                temp_list.append(1)
            else:
                temp_list.append(0)
                temp_list.append(0)
        if j != i:
            B[f'{i},{j}'] = temp_list
            messages[f'{i},{j}'] = 0.5
    
V = declare_vector(p, N)

for i in range(10):
    messages_new = {}
    messages_new = perc_log_meesage_pass(B, V, messages)
    messages = {}
    V_new = update_v_and_messages(p, messages_new)
    V = []
    V = V_new
    messages = messages_new
```
