```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt

from bp_error.bayes import posterior_loop, posterior_vectorized

# Two-urn setup
prior = jnp.array([0.5, 0.5])  # [P(A), P(B)]
likelihoods = jnp.array([
    [0.7, 0.3],  # urn A: P(W|A)=0.7, P(K|A)=0.3
    [0.3, 0.7],  # urn B
])

# Generate 100 draws from urn A
key = jax.random.PRNGKey(0)
# TODO: use jax.random.choice with probabilities [0.7, 0.3] to generate observations.
observations = jax.random.choice(key, a=2, shape=(100,), p=likelihoods[0])
# TODO: for n in 1..100, compute the posterior after the first n observations,
# and store P(A | data) in a list.
for n in range(1, 101):
    posterior = posterior_vectorized(prior, likelihoods, observations[:n])
    print(f"After {n} observations, P(A | data) = {posterior[0]:.4f}")
# TODO: plot P(A | data) vs. n.
plt.plot(range(1, 101), [posterior_vectorized(prior, likelihoods, observations[:n])[0] for n in range(1, 101)])
plt.xlabel("Number of Observations")
plt.ylabel("P(A | data)")
plt.title("Posterior Probability of Urn A Over Time")
plt.grid()
plt.show()
```
